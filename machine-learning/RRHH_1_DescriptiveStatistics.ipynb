**Step 1. Read dataset**

```
#Pandas: Dataset library
import pandas as pd

#Read dataset
rrhh_ds = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSfE_9K1uPYtXkEA_oncBZEGne6D6uhJSBCDqASxid4jIlQiyu_Z3629s0yyhNQxxkb6Q4wnUwDtlNr/pub?gid=0&single=true&output=csv")

#Print dataset
rrhh_ds.head()
```

**Step 2. It is necessary to know and understand the data**

```
#Drop IdEmpleado column
#rrhh_ds.drop(['IdEmpleado'], axis=1, inplace=True)

# Set it to None to display all columns in the dataframe
pd.set_option('display.max_columns', None)

#Similar functions for others measures: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
print("\n=====Attribute mean:=====\n")
print(rrhh_ds.mean())

print("\n=====Basic statistics:=====\n")
print(rrhh_ds.describe().transpose())

print("\n=====SatisfaccionLaboral_Disc distribution:=====\n")
st_ds = rrhh_ds.groupby('SatisfaccionLaboral_Disc')['SatisfaccionLaboral_Disc'].count()
print(st_ds)
```

**Step 3A. Descriptive analytics are very useful**

```
#Matplotlib: 2D and 3D plots
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))

#Histogram
plt.subplot(2, 2, 1)
plt.hist(rrhh_ds['SatisfaccionLaboral_Disc'])
plt.xlabel('Satisfaccion Laboral')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.hist(rrhh_ds['UltimaEvaluacion_Disc'])
plt.xlabel('Ultima Evaluacion')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.hist(rrhh_ds['ProyectosRealizados_Disc'])
plt.xlabel('Proyectos Realizados')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.hist(rrhh_ds['Antiguedad_Disc'])
plt.xlabel('Antiguedad')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()


#plt.savefig('SLHistogram.png')
```

**Step 3B. Descriptive analytics are very useful**

```
plt.figure(figsize=(12,6))

# matplotlib histogram
plt.subplot(2, 2, 1)
plt.hist(rrhh_ds['NivelSatisfaccion'], color = 'blue', edgecolor = 'black')
plt.xlabel('Satisfaccion Laboral')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.hist(rrhh_ds['UltimaEvaluacion'], color = 'green', edgecolor = 'black')
plt.xlabel('Ultima Evaluacion')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.hist(rrhh_ds['ProyectosRealizados'], color = 'red', edgecolor = 'black')
plt.xlabel('Proyectos Realizados')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.hist(rrhh_ds['HorasMensuales'], color = 'yellow', edgecolor = 'black')
plt.xlabel('Horas Mensuales')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()

```

**Step 3C. Descriptive analytics are very useful**

```
plt.figure(figsize=(12,6))

# Scatter plot
plt.subplot(1, 2, 1)
x = rrhh_ds['NivelSatisfaccion']
y = rrhh_ds['UltimaEvaluacion']
label = rrhh_ds['Renuncia_Disc']
color= ['red' if l == "SI" else 'green' for l in label]
plt.scatter(x, y, color = color)
plt.xlabel('NivelSatisfaccion')
plt.ylabel('UltimaEvaluacion')
        
plt.subplot(1, 2, 2)
x = rrhh_ds['HorasMensuales']
y = rrhh_ds['Antiguedad']
label = rrhh_ds['Renuncia_Disc']
color= ['red' if l == "SI" else 'green' for l in label]
plt.scatter(x, y, color = color)
plt.xlabel('HorasMensuales')
plt.ylabel('Antiguedad')

plt.show()
```

**Step 3D. Descriptive analytics are very useful**

```
#Correlation Matrix   
corr = rrhh_ds.corr()

#print("\nCorrelation matrix:\n")
#print(corr)

#Correlation plot
fig = plt.figure(figsize=(7, 7))
plt.matshow(corr, fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns,fontsize=7, rotation=45);
plt.yticks(range(len(corr.columns)), corr.columns,fontsize=7, rotation=45);
cb = plt.colorbar()
cb.ax.tick_params(labelsize=8)
plt.show()        
```

**A 3D Projection Example**

```
#A simple 3D chart
serie1 = rrhh_ds.groupby('SatisfaccionLaboral_Disc')['IdEmpleado'].count().to_frame().reset_index(level=['SatisfaccionLaboral_Disc'])
#serie1['SatisfaccionLaboral_Disc']
#serie1['IdEmpleado']
#print(len(serie1))

serie2 = rrhh_ds.groupby('UltimaEvaluacion_Disc')['IdEmpleado'].count().to_frame().reset_index(level=['UltimaEvaluacion_Disc'])

from mpl_toolkits.mplot3d import Axes3D 
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g']
yticks = [1, 2]

xs = np.arange(len(serie1))
ys = serie1['IdEmpleado']
cs = ['r'] * len(xs)
cs[0] = 'c'

ax.bar(xs, ys, zs=1, zdir='y', color=cs, alpha=0.8)

xs = np.arange(len(serie2))
ys = serie2['IdEmpleado']
cs = ['g'] * len(xs)
cs[0] = 'c'

ax.bar(xs, ys, zs=2, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_yticks(yticks)

plt.show()
```
